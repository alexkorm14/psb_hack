```python
# standard libraries
import sys
import os
import json
import re
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
# ds libraries
import pandas as pd
import numpy as np
import joblib
import gc
# свои функции по оценке метрик
path_to_model_utils = '/content/drive/MyDrive/psb_hack'
sys.path.append(os.path.join(path_to_model_utils,'model_utils'))
from features_pack import *
```

## Загрузка данных

```python
# чтение данных
dates = ['Дата бронирования','Дата отмены', 'Заезд', 'Выезд']
train_full = pd.read_csv(os.path.join(path_to_folder,'data','full_train.csv'),parse_dates=dates)
train = pd.read_csv(os.path.join(path_to_folder,'data','train.csv'),parse_dates=dates)
oot_val = pd.read_csv(os.path.join(path_to_folder,'data','oot_val.csv'),parse_dates=dates)
oof_val = pd.read_csv(os.path.join(path_to_folder,'data','oof_val.csv'),parse_dates=dates)
test_data = pd.read_csv(os.path.join(path_to_folder,'data','test.csv'),parse_dates=list(filter(lambda x: x != 'Дата отмены',dates)))

gc.collect()
```

## Features pack 1

```python
# расчет фичей
train_full_v1 = features_pack_v1(train_full)
train_v1 = features_pack_v1(train)
oot_val_v1 = features_pack_v1(oot_val)
oof_val_v1 = features_pack_v1(oof_val)
test_data_v1 = features_pack_v1(test_data)
# сохранение фичей
# сохраним все в csv, чтобы быстрее было считывать
train_full_v1.to_csv(os.path.join(path_to_folder,'data','full_train_features_pack_v1.csv'),header = True,index=False)
train_v1.to_csv(os.path.join(path_to_folder,'data','train_features_pack_v1.csv'),header = True,index=False)
oot_val_v1.to_csv(os.path.join(path_to_folder,'data','oot_val_features_pack_v1.csv'),header = True,index=False)
oof_val_v1.to_csv(os.path.join(path_to_folder,'data','oof_val_features_pack_v1.csv'),header = True,index=False)
test_data_v1.to_csv(os.path.join(path_to_folder,'data','test_features_pack_v1.csv'),header = True,index=False)
```
